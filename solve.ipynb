```kotlin
// Функция утилита
fun Double.round(decimals: Int): Double {
    var multiplier = 1.0
    repeat(decimals) { multiplier *= 10 }
    return round(this * multiplier) / multiplier
}
```

```kotlin
val df = DataFrame.readCsv("./data/input.csv").add("D"){
    ("left"<Double>() - "right"<Double>()).round(3)
}.add("r"){
    ("D"<Double>() / 2.0).round(3)
}
df
```

```kotlin
val lambda = 0.66 * 10.0.pow(-3)
```

```kotlin
val pairs = mapOf(1 to 7, 2 to 8, 3 to 9, 7 to 1, 8 to 2, 9 to 3)
```

```kotlin
// Представленные значение в мм (10^(-3))

val results = df.add("rk + rm" ) {
    if (N in pairs.keys){
        (df.first { row -> row.N == pairs[it.N] }.get { "r"<Double>() } + "r"<Double>()).round(3) // 10^-3
    }else {
        return@add "-"
    }
}.add("rk - rm") {
    if (N in pairs.keys){
        (df.first { row -> row.N == pairs[it.N] }.get { "r"<Double>() } - "r"<Double>()).round(3).absoluteValue
    }else {
        return@add "-"
    }
}.add("R") {
    val n = "N"<Int?>()!!
    if (n in pairs.keys){
        return@add (("rk + rm"<Double?>()!! * "rk - rm"<Double?>()!!)/((pairs[n]!! - n) * lambda)).roundToInt().absoluteValue
    }else{
        return@add "-"
    }
}
results
```

```kotlin
val Ravg = results.take(3).mean { R.convertToDouble() }
Ravg.round(2)
```

### Вычисление погрешностей
#### Систематическая погрешность



$$
\begin{align*}
\theta_R
&= \left|\frac{\partial R}{\partial r_k}\right|\theta_{r_k}
+ \left|\frac{\partial R}{\partial r_m}\right|\theta_{r_m} \\[6pt]
&= \frac{2r_k\,\theta_{r_k}}{\lambda\,(k - m)}
+ \frac{\lvert -2r_m\rvert\,\theta_{r_m}}{\lambda\,(k - m)} \\[6pt]
&= \frac{2\,(r_k + r_m)}{\lambda\,(k - m)}\,\theta_r.
\end{align*}
$$



```kotlin
val thetaR = 0.005
```

```kotlin
val resultsWithTheta = results.update { R }.where { R.toString() != "-" }.with {
    val R = (R as Int)
    val sum = "rk + rm"<Double>()
    val theta = (2 * sum)/(lambda * (N - pairs[N]!!.toInt())) * thetaR
    return@with "$R plus.minus ${theta.roundToInt().absoluteValue}"
}

resultsWithTheta.apply { writeCsv("./data/output.csv") }
```
